```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings. filterwarnings("ignore")
```

```python
data=pd.read_csv("Automobile_data.csv")
```

```python
data
```

```python
data.info()
```

```python
# We need to check the unique data of all columns so as to make sure is there any characters or blank value in the data.

for col in data:
    print(col,":",data[col].unique())
    print("******************************************")
```

```python
# Replacing the question mark(?) with Nan values.

# We can use replace function only on string data not on numbers and Nan values.

data["normalized-losses"].replace("?",np.nan,inplace=True)
data["horsepower"].replace("?",np.nan,inplace=True)
```

```python
col=["normalized-losses","horsepower"]

for i in col:
    print(i,":",data[i].unique())
    print("************************************")
```

```python
data.info()
```

```python
data.isnull().sum() # This function will give number of Nan values in particular column.
```

```python
# Changing the data type of normalized losses & horsepower.

# We are going to change data type of normalized losses & horsepower as it in object.
# Where there are Nan values in any column we cant directly change the column data type to interger.
# We have to change it by float value first.

data["normalized-losses"]=data["normalized-losses"].astype("float64")
data["horsepower"]=data["horsepower"].astype("float64")
```

```python
data.info()
```

```python
data.describe()
```

```python
# As we have checked the mean value of normalized-losses and horsepower.
# We can replace Nan values with mean.

# Filna function is used to fill null values with any numbers.

data["normalized-losses"].fillna(122,inplace=True)
data["horsepower"].fillna(104,inplace=True)
```

```python
data.info()
```

```python
data.describe()
```

```python
# Handling outliers.

# An outliers is an observation that lies an abnormal distance from other values in a column.

plt.figure()
sns.boxplot(data["price"])
plt.show()
```

```python
# We are going to find Q1,Q2,Q3 & IQR(Inter Quartile Range).

Q1=np.quantile(data["price"],0.25) # 25% of data
Q2=np.quantile(data["price"],0.50) # 50% of data
Q3=np.quantile(data["price"],0.75) # 75% of data

IQR=Q3-Q1

print("Q1:",Q1)
print("Q2:",Q2)
print("Q3:",Q3)
print("IQR:",IQR)
```

```python
# Upper whisker & Lower whisker.

UW=Q3+(1.5*IQR)
LW=Q1-(1.5*IQR)

print("LW:",LW)
print("UW:",UW)

# Any values less than Lower whisker and more than Upper whisker is considered as outliers.
```

```python
# Removing all outliers, we are going to plot makers with price.

plt.figure(figsize=(12,7))
sns.boxplot(data=data,x="price",y="make",palette="tab10")
plt.show()
```

```python
# Removing all outliers with loop.

for maker in data["make"].unique(): # for loop will run on maker column of each maker as we given unique.
    price=data[data["make"]==maker]["price"] # All the maker price data will be store in price variable with index.
    Q1=np.quantile(price,0.25) # finding Q1.
    Q3=np.quantile(price,0.75) # finding Q3.
    IQR=Q3-Q1 # finding IQR for upper whsiker & Lower whisker
    UW=Q3+(1.5*IQR) # upper whsiker
    LW=Q1-(1.5*IQR) # lower whisker
    
    i=price[price>UW].index # All the data of price will be get stored in i with index number if price > UW
    if len(i)!=0: # i list length should not have 0 elements.It will store all row(index)number which is > UW
        data.drop(i,inplace=True)
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(data=data,x="price",y="make",palette="tab10")
plt.show()
```
